# Part1: Preprocessing

```python
#Reading all of the csv files 
data10 = pd.read_csv("input_folder/Events_2010.csv")
data11 = pd.read_csv("input_folder/Events_2011.csv")
data12 = pd.read_csv("input_folder/Events_2012.csv")
data13 = pd.read_csv("input_folder/Events_2013.csv")
data14 = pd.read_csv("input_folder/Events_2014.csv")
data15 = pd.read_csv("input_folder/Events_2015.csv")
data16 = pd.read_csv("input_folder/Events_2016.csv")
data17 = pd.read_csv("input_folder/Events_2017.csv")
data18 = pd.read_csv("input_folder/Events_2018.csv")

df2=pd.read_csv('input_folder/RegularSeasonDetailedResults.csv')

ranking=pd.read_csv('input_folder/MasseyOrdinals.csv')
```

```python
data10=data10[(data10['WPoints']>0) | (data10['LPoints']>0)]
data11=data11[(data11['WPoints']>0) | (data11['LPoints']>0)]
data12=data12[(data12['WPoints']>0) | (data12['LPoints']>0)]
data13=data13[(data13['WPoints']>0) | (data13['LPoints']>0)]
data14=data14[(data14['WPoints']>0) | (data14['LPoints']>0)]
data15=data15[(data15['WPoints']>0) | (data15['LPoints']>0)]
data16=data16[(data16['WPoints']>0) | (data16['LPoints']>0)]
data17=data17[(data17['WPoints']>0) | (data17['LPoints']>0)]
data18=data18[(data18['WPoints']>0) | (data18['LPoints']>0)]

frames=[data10,data11,data12,data13,data14,data15,data16,data17,data18]
data=pd.concat(frames)

data.reset_index(inplace = True, drop = True)
```

```python

data['matchUniqueID']=data['Season'].map(str)+"_"+data['DayNum'].map(str)+"_"+data['WTeamID'].map(str)+"_"+data['LTeamID'].map(str)
df2['matchUniqueID']=df2['Season'].map(str)+"_"+df2['DayNum'].map(str)+"_"+df2['WTeamID'].map(str)+"_"+df2['LTeamID'].map(str)

df3=df2[['matchUniqueID','WLoc']]
df6=pd.merge(data,df3,on='matchUniqueID',how='inner')
df6
```

```python
def func1(df):
        if df['WTeamID']<df['LTeamID']:
            return df['WTeamID']
        else:
            return df['LTeamID']
def func2(df):
        if df['WTeamID']>df['LTeamID']:
            return df['WTeamID']
        else:
            return df['LTeamID']
def func3(df):
        if df['WTeamID']<df['LTeamID']:
            return df['WPoints']
        else:
            return df['LPoints']
def func4(df):
        if df['WTeamID']>df['LTeamID']:
            return df['WPoints']
        else:
            return df['LPoints']
def func5(df):
        if df['WTeamID']>df['LTeamID'] and df['WTeamID']==df['EventTeamID']:
            return 0
        elif df['WTeamID']<df['LTeamID'] and df['WTeamID']==df['EventTeamID']:
            return 1
        elif df['WTeamID']>df['LTeamID'] and df['LTeamID']==df['EventTeamID']:
            return 1
        else:
            return 0
def func6(df):
        if df['WTeamID']>df['LTeamID']:
            return 0
        else:
            return 1

def func7(df):
        if df['WTeamID']<df['LTeamID'] and df['WLoc']=='H':
            return 1
        elif df['WTeamID']>df['LTeamID'] and df['WLoc']=='A':
            return 1
        else:
            return 0
def func8(df):
        if df['WTeamID']<df['LTeamID'] and df['WLoc']=='A':
            return 1
        elif df['WTeamID']>df['LTeamID'] and df['WLoc']=='H':
            return 1
        else:
            return 0
        
```

```python
df7=df6.copy()
df7['TeamID1']=df7.apply(func1,axis=1)
df7['TeamID2']=df7.apply(func2,axis=1)
df7['Points1']=df7.apply(func3,axis=1)
df7['Points2']=df7.apply(func4,axis=1)
df7['EventTeam']=df7.apply(func5,axis=1) #If the event is  of Team1 then 1 else 0
df7['Win1']=df7.apply(func6,axis=1) # Whether Team1 wins
df7['isTeam1Home']=df7.apply(func7,axis=1) #Is Team1 is playing in Home Field
df7['isTeam2Home']=df7.apply(func8,axis=1) #Is Team2 is playing in Home Field ; Both cannot be 1; also if both are 0 then
#they are playing in neutral field 
```

# Part 3 : Training and Testing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import tensorflow as tf
```

```python
# If you want to start directly from part2 then uncomment the below cell and start and get df10
# run the above cell to import libraries
```

```python
df10=pd.read_csv('input_folder/inputWithRank.csv')
df10.drop(labels='Unnamed: 0',axis=1,inplace=True)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df10['EventType']= label_encoder.fit_transform(df10['EventType'])
df10.drop(labels='matchUniqueID',axis=1,inplace=True)
df10.drop(labels=['DayNum','TeamID1','TeamID2'],axis=1,inplace=True)
```

```python

model2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=[9,]),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

df11=df10[df10['Season']<=2017]
df12=df10[df10['Season']>2017]

model2.compile(optimizer=tf.keras.optimizers.RMSprop(1e-3), loss='binary_crossentropy', metrics=['accuracy', 'AUC'])
model2.summary()
model2.fit(df11[['isTeam1Home', 'isTeam2Home', 'Points1', 'Points2','ElapsedSeconds', 'EventTeam', 'EventType', 'OrdRankTeam1','OrdRankTeam2']].to_numpy(),
          df11['Win1'].to_numpy(),
          verbose=1,
          shuffle=True,
          epochs=20,
          batch_size=512)
```

```python
y_pred=model2.predict(df12[['isTeam1Home', 'isTeam2Home', 'Points1', 'Points2','ElapsedSeconds', 'EventTeam', 'EventType', 'OrdRankTeam1','OrdRankTeam2']])
y_pred=pd.DataFrame(y_pred)
y=df12[['Win1']]
y.insert(1,'y_pred',y_pred)
y.rename(columns={'Win1':'y_true'},inplace=True)
y['Error']=abs(y['y_true']-y['y_pred'])**2


y['y_pred_binary']=[1 if x>=0.5 else 0 for x in y['y_pred']]
y['Misclassification']=(y['y_pred_binary']-y['y_true'])**2  # here 1=misclassification, 0 means correct prediction
print("Mean Square Error is :",y['Error'].mean())


print("Root Mean Square Error RMSE is :",np.sqrt(y['Error'].mean()))


from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
r2 = r2_score(y['y_true'], y['y_pred'])
print('R2 value is', r2)

accuracy=accuracy_score(y['y_true'], y['y_pred_binary'])
print("Accuracy is", accuracy)

print("Misclassification Error is", y['Misclassification'].mean())
```
